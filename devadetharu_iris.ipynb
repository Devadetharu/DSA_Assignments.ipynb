```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris_new = iris.select_dtypes(include=['number'])
corr = iris_new.corr()
corr
```

```python
sns.violinplot(iris) # combines boxplot + density and the use is distribution and density
```

```python
sns.stripplot(iris) # shows individual data points and the use is see exact values
```

```python
sns.swarmplot(iris) # like stripplot, but no overlap and the use is clear individual points
```

```python
sns.regplot(data=iris,x='sepal_length',y='sepal_width') # scatter + regression line and the use is linear relationship
```

```python
sns.pointplot(iris) #shows mean with confidence interval and the use is compare averages
```

```python
sns.lmplot(data=iris,x='sepal_length',y='sepal_width') #regression plot with categories and the use regression with groups
```

```python
sns.jointplot(iris) #scatter + distributions and the use is relationship + histogram
```

```python
sns.clustermap(iris_new) # heatmap with clustering and the use is group similar data
```
