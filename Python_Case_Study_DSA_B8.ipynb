# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(list_numbers):
  minimum_value = list_numbers[0]
  # setting the first value of  the list as minimum value
  maximum_value = list_numbers[0]
  # setting the first value of  the list as maximum value

  for numbers in list_numbers:
  # Looping through the entire list to check for
  # maximum value and minimum value
    if(numbers >= maximum_value):
      maximum_value = numbers
    if(numbers <= minimum_value):
      minimum_value = numbers

  return(minimum_value, maximum_value)


list_numbers = [12,31,45,23,56,7]
range = find_range(list_numbers)
print(range)
```
